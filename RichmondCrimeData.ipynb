```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data2)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Extract the month from the 'Date' column
df['Month'] = df['Date'].dt.month

# Group the data by month and count the number of crimes in each month
monthly_crime_counts = df['Month'].value_counts().sort_index()

# Define the labels for the x-axis (months)
month_labels = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(month_labels, monthly_crime_counts)
plt.xlabel("Month")
plt.ylabel("Number of Crimes")
plt.title("Number of Crimes by Month")
plt.show()
```

```python
import matplotlib.pyplot as plt
crime_by_month = data2.groupby(['Month', 'Description']).size().unstack(fill_value=0)

# Plotting
crime_by_month.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Crime Frequency by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.legend(title='Crime Description', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Set your Mapbox API token
px.set_mapbox_access_token('YOUR_API_TOKEN')

data2 = pd.DataFrame(data2)

# Create a Scattermapbox plot using Plotly Express
fig = px.scatter_mapbox(
    data2,
    lat="Y Coordinate",
    lon="X Coordinate",
    hover_name="Location Description",
)

# Customize the map layout
fig.update_layout(
    mapbox_style="carto-positron",
    mapbox_zoom=12,  # Adjust the initial zoom level
    mapbox_center={"lat": data2["Y Coordinate"].mean(), "lon": data2["X Coordinate"].mean()},  # Set the center of the map
    title="Crime Locations",
)

# Show the plot
fig.show()
```

```python
crime_by_month = dfc.groupby(['Month', 'Crime']).size().unstack(fill_value=0)
crime_by_month
```

```python
import matplotlib.pyplot as plt
crime_by_month = dfc.groupby(['Month', 'Crime']).size().unstack(fill_value=0)

# Plotting
crime_by_month.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Crime Frequency by Month')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.legend(title='Crime Description', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
