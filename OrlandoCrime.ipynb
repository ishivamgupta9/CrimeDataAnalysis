```python
import pandas as pd

file_id="1S_G1rUEQOaTljrVS10qI2j_bcXs2VmWA"
file_url = f"https://drive.google.com/uc?id={file_id}"

data = pd.read_csv(file_url)

data
```

```python
dropcolumn = ["Review Letter", "Fatal (Officer)", "Officer Hit", "Fatal", "Suspect Hit", "Officer Tenure", "Officer Age", "Ethnicity", "Officer Race", "Officer Gender"]
data.drop(dropcolumn, axis=1, inplace=True)
```

```python
data.isna().any()
```

```python
data.drop(['Number of Officers Involved','Officer Name','Suspect Race','Suspect Gender'],axis=1,inplace=True)
```

```python
data
```

```python
data['Georeferenced Column']
```

```python
# Extracting latitude and longitude from the 'Georeferenced Column'
data[['Longitude', 'Latitude']] = data['Georeferenced Column'].str.extract(r'POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)').astype(float)

data
```

```python
data["Date"]=pd.to_datetime(data["Date"])
data["Month"]=data["Date"].dt.strftime('%B')
```

```python
data
```

```python
import matplotlib.pyplot as plt
import calendar

crime_by_month=data.groupby(["Month","Incident Type"]).size().unstack(fill_value=0)


month_order = [calendar.month_name[i] for i in range(1, 13)]  # Create a list of month names
crime_by_month = crime_by_month.reindex(month_order, axis=0)

crime_by_month.plot(kind='line', stacked=True, figsize=(12, 6))
plt.title('Crime Frequency by Month')
plt.xlabel("Month")
plt.ylabel("Frequency")
plt.legend(title="Crime Description", bbox_to_anchor=(1.05,1),loc='best')
plt.show()
```
